```python
import re

#提取文件中的命令
def parse_data(inputdir):
    dataset = []
    #匹配任何非单词字符 
    regEx = re.compile('\\W*')
    with open(inputdir,encoding='UTF-8') as f:
        data_tem = []
        for line in f.readlines():
            tem = []
            #去掉行尾换行符
            line=line.strip('\n')
            #按照空格划分字符串
            listoftoken = re.split(' ', line)            
            for token in listoftoken:
                #去除空格
                if len(token)>0:
                    tem.append(token)
                #去除行号
            data_tem.append(tem[1])
        dataset = data_tem
        return dataset
```
