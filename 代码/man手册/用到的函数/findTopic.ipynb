```python
import os
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import make_pipeline

document = []
name = []
topicVector = {}

stopword = []

with open('C:\\Users\\Ricky\\Desktop\\work\\stopwords.txt') as f:
    for i in range(891):
        stopword.append(f.readline().strip('\n'))

def findTopic(rootdir,stopword,model='lsa',ngram_range=(1,2),dimension=50):
    
    #列出文件夹下所有文件
    fileList = os.listdir(rootdir)
    
    #取出每一个命令名字
    for i in fileList:
        name.append(i[0:-4])
    
    #链接目录地址和每一个文件名，得到每个文件地址，把所有文档综合起来
    for i in range(0,len(fileList)):
        path = os.path.join(rootdir,fileList[i])
        if os.path.isfile(path):
            with open(path,'r+',encoding='utf-8') as f:
                document.append(f.read())
    
    #把文档输入到模型里训练
    if model == 'lsa':
        #设置停用词
        vec = TfidfVectorizer(ngram_range,stop_words=stopword)
        X = vec.fit_transform(document).toarray()

        #取多少维作为主题特征向量
        svd = TruncatedSVD(dimension)
        norm = Normalizer(copy=False)
        lsa = make_pipeline(svd, norm)
        allTopicVector = lsa.fit_transform(X)

    for i in range(0,len(fileList)):
        topicVector[name[i]] = allTopicVector[i]
        

    return topicVector
        
```

```python
a = findTopic('C:\\Users\\Ricky\\Desktop\\work\\txt23')
```

```python
x = []
for topic in a.values():
    x.append(topic)
```

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
```

```python
def Model_DBSCAN(dataset,best_epsilon=0.1,best_num=3):
    '''
    使用DBSCAN聚类结果为数据贴标签
    '''
    X = dataset
    
    clst = DBSCAN(eps = best_epsilon, min_samples = best_num)
    clst.fit(X)
    clst_labels = clst.labels_
    if len(set(clst_labels))>1:
        evalue=metrics.silhouette_score(X,clst.labels_,metric='euclidean')
    else:
        evalue="no exception people"
    #输出评价系数
    print ("Evaluate Ratio: %s" % evalue)
    print ("=============================================")
    for clst_lab in set(clst_labels):
        print ("Number of the %s class: %s" % (clst_lab,list(clst_labels).count(clst_lab)))
    print( "=============================================")
    print( "Number of the labels: %s" % len(clst_labels))
    
    return clst_labels
```
